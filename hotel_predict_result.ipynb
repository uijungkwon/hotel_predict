#### 데이터 전처리

```python
#범주형 데이터를 수치형데이터로 변환

train['room_type_reserved'] = train['room_type_reserved'].astype('category').cat.codes
test['room_type_reserved'] = test['room_type_reserved'].astype('category').cat.codes

train['market_segment_type'] = train['market_segment_type'].astype('category').cat.codes
test['market_segment_type'] = test['market_segment_type'].astype('category').cat.codes


train['type_of_meal_plan'] = train['type_of_meal_plan'].astype('category').cat.codes
test['type_of_meal_plan'] = test['type_of_meal_plan'].astype('category').cat.codes

train['lead_time'] = pd.qcut(train['lead_time'], 5)
test['lead_time'] = pd.qcut(test['lead_time'], 5)

train['lead_time'] = train['lead_time'].astype('category').cat.codes
test['lead_time'] = test['lead_time'].astype('category').cat.codes
```

```python
#target feature 수치형 데이터로 변환
train['booking_status'] = train['booking_status'].replace({'Canceled':1, 'Not_Canceled':0})
```

```python
#이상치를 계산하는 함수 작성 
def outlier_iqr(data, column): 

    # lower, upper 글로벌 변수 선언하기     
    global lower, upper    
    
    # 4분위수 기준 지정하기     
    q25, q75 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)          
    
    # IQR 계산하기     
    iqr = q75 - q25    
    
    # outlier cutoff 계산하기     
    cut_off = iqr * 1.5          
    
    # lower와 upper bound 값 구하기     
    lower, upper = q25 - cut_off, q75 + cut_off     
    
    print('IQR은',iqr, '이다.')     
    print('lower bound 값은', lower, '이다.')     
    print('upper bound 값은', upper, '이다.')    
    
    # 1사 분위와 4사 분위에 속해있는 데이터 각각 저장하기     
    data1 = data[data[column] > upper]     
    data2 = data[data[column] < lower]    
    
    # 이상치 총 개수 구하기
    return print(column, '총 이상치 개수는', data1.shape[0] + data2.shape[0], '이다.')
```

```python
#train 데이터의 전체 column의 이상치 확인 -> 이상치가 1000개 이상인 column은 feature 선택에서 제외
outlier_iqr(train,'no_of_adults') 
outlier_iqr(train,'no_of_children')
outlier_iqr(train,'no_of_weekend_nights')
outlier_iqr(train,'no_of_week_nights')
outlier_iqr(train,'type_of_meal_plan') 
outlier_iqr(train,'required_car_parking_space')
outlier_iqr(train,'room_type_reserved') 
outlier_iqr(train,'arrival_year')      
outlier_iqr(train,'arrival_month')
outlier_iqr(train,'arrival_date')
outlier_iqr(train,'market_segment_type') 
outlier_iqr(train,'repeated_guest')
outlier_iqr(train,'no_of_previous_cancellations') 
outlier_iqr(train,'no_of_previous_bookings_not_canceled') 
outlier_iqr(train,'avg_price_per_room') 
outlier_iqr(train,'no_of_special_requests')
outlier_iqr(train,'lead_time')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
#train의 수치형 칼럼에 대해 상관계수 구하기
df = train.iloc[ :, 1:-1]
```

```python
#train의 전체 칼럼에 대해 상관계수 구하기
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))    
sns.heatmap(df.corr(), annot=True)
```

```python
##feature 선택

feature_names = [
#'no_of_adults',
'no_of_children',
#'type_of_meal_plan',
##'no_of_weekend_nights'                    ,
##'no_of_week_nights'                       ,
#'required_car_parking_space'              ,
#'arrival_year'                            ,
#'arrival_month'                         ,
#'arrival_date',
#'room_type_reserved',
'lead_time'                               ,
#'repeated_guest'                          ,
'no_of_previous_cancellations'            ,
'no_of_previous_bookings_not_canceled'    ,
'avg_price_per_room'                      ,
'no_of_special_requests',
'market_segment_type',
    
]
X = train[feature_names]
y = train["booking_status"]
```

```python
#상관계수가 높은 feature들을 선택하기 위해 한번 더 시각화
plt.figure(figsize=(10, 10))    
sns.heatmap(train[feature_names].corr(), annot=True)
```

#### Decision classifier 모델 생성 및 예측

```python
###Decision Tree 사용 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=11)
dt_model = DecisionTreeClassifier(max_depth=10 ,random_state=156)
dt_model.fit(X_train,y_train)
dt_pred = dt_model.predict(X_test)
print('DecisionTreeClassifier 정확도: {0:.4f}'.format(accuracy_score(y_test, dt_pred)))
```
